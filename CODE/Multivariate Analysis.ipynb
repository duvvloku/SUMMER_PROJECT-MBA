```python
import pandas as pd 
import numpy as np
from IPython.display import display, HTML
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)

import matplotlib.pyplot as plt   
          #importing seaborn for statistical plots
import seaborn as sns
          # To enable plotting graphs in Jupyter notebook
%matplotlib inline
          #break the X and y dataframes into training set and test set. For this we will use
          #Sklearn package's data splitting function which is based on random function
```

```python
data = pd.read_csv('learning_style_dataset_stats.csv')
display(data.head())
```

```python
s = data.describe()
display(HTML(s.to_html()))
```

```python
s.to_csv('describe.csv',index = True)
```

```python
s = data.corr()
display(HTML(s.to_html()))
```

```python
s.to_csv('correlation.csv',index = True)
```

```python
fig,ax = plt.subplots(figsize=(10, 10))   
sns.heatmap(data.corr(), ax=ax, annot=True, linewidths=0.05, fmt= '.2f',cmap="magma")
plt.show()
```

```python
s = data.cov()
display(HTML(s.to_html()))
```

```python
fig,ax = plt.subplots(figsize=(10, 10))   
sns.heatmap(data.cov(), ax=ax, annot=True, linewidths=0.05, fmt= '.2f',cmap="GnBu")
plt.show()
```

```python
s = data.kurtosis().to_frame()
display(HTML(s.to_html()))
```

```python
s = data.skew().to_frame()
display(HTML(s.to_html()))
```

```python
s = data.var().to_frame()
display(HTML(s.to_html()))
```

```python
sns.pairplot(data,diag_kind='kde')
```
