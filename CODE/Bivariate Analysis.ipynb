```python
import scipy.stats as stats

f_stat, p_value = stats.f_oneway(data['V1'][data['style'] == 'V'],
               data['V1'][data['style'] == 'A'],
               data['V1'][data['style'] == 'R'],
               data['V1'][data['style'] == 'K'],
               data['V1'][data['style'] == 'Multi'])

Ho = "V1 has no effect on learning_style"   # Stating the Null Hypothesis
Ha = "V1 has an effect on learning_style"   # Stating the Alternate Hypothesis

print('f_statistic : ',f_stat)
print('')
print('p_value : ',p_value)
print('')

if p_value < 0.05:  # Setting our significance level at 5%
    print(f'{Ha} as the p_value ({p_value.round(3)}) < 0.05')
else:
    print(f'{Ho} as the p_value ({p_value.round(3)}) > 0.05')
```

```python
results = pd.DataFrame({'Variable':['V1'],
                        'F statistic': f_stat,
                       'p value':p_value,},index={'1'})
results = results[['Variable','F statistic', 'p value']]
results
```

## categorical variables

```python
import statsmodels.api as sm
import scipy.stats as stats
```

```python
Ho = "Gender has no effect on learning_style"   # Stating the Null Hypothesis
Ha = "Gender has an effect on learning_style"   # Stating the Alternate Hypothesis

crosstab = pd.crosstab(data['Gender'],data['learning_style'])

chi, p_value, dof, expected =  stats.chi2_contingency(crosstab)

display(crosstab)
print('')
print('p_value =',p_value)
print('')
print('chisquare value =',chi)
print('')
print('degrees of freedon =',dof)
print('')
print("Expected crosstab:")
display(pd.DataFrame(expected))
print('')

if p_value < 0.05:  # Setting our significance level at 5%
    print(f'{Ha} as the p_value ({p_value.round(3)}) < 0.05')
else:
    print(f'{Ho} as the p_value ({p_value.round(3)}) > 0.05')
```

```python
results = pd.DataFrame({'Variable':['Gender'],
                        'chisquare value': chi,
                       'p value':p_value,},index={'1'})
results = results[['Variable','chisquare value', 'p value']]
results
```

```python
g = sns.catplot(x ='Gender', hue ='learning_style', data = data, kind ='count')
g.fig.set_figwidth(10)
g.fig.set_figheight(5)
plt.xticks(rotation='vertical')
plt.show()
```

```python
data.groupby('Gender')['learning_style'].value_counts().groupby(level=0).head().to_frame()
```
