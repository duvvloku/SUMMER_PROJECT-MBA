```python
import pandas as pd 
import numpy as np
from IPython.display import display, HTML
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)

import matplotlib.pyplot as plt   
          #importing seaborn for statistical plots
import seaborn as sns
          # To enable plotting graphs in Jupyter notebook
%matplotlib inline
          #break the X and y dataframes into training set and test set. For this we will use
          #Sklearn package's data splitting function which is based on random function
```

```python
data = pd.read_csv('learning_style_dataset_stats.csv')
display(data.head())
```

```python
plt.figure(figsize=(15,8))
sns.countplot(data['learning_style'])
```

```python
s = data.learning_style
counts = s.value_counts()
percent = s.value_counts(normalize=True)
percent100 = s.value_counts(normalize=True).mul(100).round(2).astype(str) + '%'
cum_percent = (100*(counts.cumsum() / counts.sum())).round(2).astype(str) + '%' 
learning_style_df = pd.DataFrame({'frequency': counts, 'percentage': percent100, 'cum_percentage':cum_percent})
```

```python
display(learning_style_df)
learning_style_df.to_csv('learning_style_df.csv',index = True)
```
